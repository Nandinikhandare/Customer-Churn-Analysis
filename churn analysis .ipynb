```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Prepare data
senior_churn = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Convert to percentage
senior_churn_percent = senior_churn.div(senior_churn.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(5,4))

bottoms = [0] * len(senior_churn_percent)

for churn_status in senior_churn_percent.columns:
    ax.bar(
        senior_churn_percent.index,
        senior_churn_percent[churn_status],
        bottom=bottoms,
        label=churn_status
    )
    bottoms = bottoms + senior_churn_percent[churn_status]

# Add percentage labels
for idx, (i, row) in enumerate(senior_churn_percent.iterrows()):
    cumulative = 0
    for churn_status in senior_churn_percent.columns:
        value = row[churn_status]
        if value > 5:  # Only label if the section is big enough
            ax.text(
                idx, 
                cumulative + value/2, 
                f'{value:.1f}%', 
                ha='center', 
                va='center', 
                color='white', 
                fontsize=9
            )
        cumulative += value

ax.set_title('Churn by SeniorCitizen (Percentage)')
ax.set_ylabel('Percentage')
ax.set_xlabel('SeniorCitizen')
ax.legend(title='Churn')
plt.xticks([0, 1], ['No', 'Yes'])  # Assuming 0 = No, 1 = Yes
plt.ylim(0, 100)
plt.show()
```

```python


# Let's say your DataFrame is called 'df'
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplots
fig, axes = plt.subplots(3, 3, figsize=(18, 12))  # 3 rows x 3 cols
axes = axes.flatten()  # Flatten to easily iterate

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue= df["Churn"])
    axes[i].set_title(f'Countplot of {col}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate labels if needed

plt.tight_layout()
plt.show()
```

#The majority of customers who do not churn tend to have serivces like PhoneService,InternetService(partcularly DSL), and OnlineSecurity enabled.
For Services like OnlineBackup,TechSupport,StreamingTV, churn rates are noticeably higher when these services are not use or unavailable.
